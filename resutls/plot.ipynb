```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
vr_acc = np.load('accs.npy')
vr_losses = np.load('losses.npy')
vr_grad_norms = np.load('grad_norms.npy')
acc = np.load('no_vr_accs.npy')
losses = np.load('no_vr_losses.npy')
grad_norms = np.load('no_vr_grad_norms.npy')
```

```python
fig, ax = plt.subplots(1, 3,  figsize=(21, 7))
fig.suptitle('MARINA convergence results')
X = np.arange(2500, step=50)
ax[0].plot(X, acc, label='MARINA')
ax[0].plot(X, vr_acc, label='VR-MARINA')
ax[0].legend()
ax[0].set_xlabel('Step')
ax[0].set_ylabel('Accuaracy')

ax[1].plot(X, losses, label='MARINA')
ax[1].plot(X, vr_losses, label='VR-MARINA')
ax[1].set_xlabel('Step')
ax[1].set_ylabel('Loss')
ax[1].legend()

ax[2].plot(X, grad_norms, label='MARINA')
ax[2].plot(X, vr_grad_norms, label='VR-MARINA')
ax[2].set_ylabel("$\| \\nabla f(x)\|^2$")
ax[2].set_xlabel('Step')
ax[2].legend()

ax[0].grid()
ax[1].grid()
ax[2].grid()
plt.savefig('compare.png')
```
