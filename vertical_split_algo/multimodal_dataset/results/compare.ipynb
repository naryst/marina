```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
train = np.load('train.npy')
train_frozen = np.load('train_frozen.npy')
```

```python
import matplotlib.pyplot as plt
# Smoothing function using a moving average
def moving_average(data, window_size):
    weights = np.repeat(1.0, window_size) / window_size
    return np.convolve(data, weights, 'valid')

# Define the window size for the moving average
window_size = 10

# Smooth the loss data using the moving average
smoothed_loss = moving_average(train, window_size)
smoothed_loss_frozen = moving_average(train, window_size)



plt.figure(figsize=(15, 30))
fig, axes = plt.subplots(1, 2, figsize=(17, 8))
axes[0].plot(train, alpha=0.75)
axes[0].plot(smoothed_loss)
axes[1].plot(train_frozen, alpha=0.75)
axes[1].plot(smoothed_loss_frozen)
axes[0].grid()
axes[1].grid()

axes[0].set_title('Loss plot')
axes[0].set_xlabel('Model step')
axes[0].set_ylabel('Loss value')

axes[1].set_title('Loss plot frozen')
axes[1].set_xlabel('Model step')
axes[1].set_ylabel('Loss value')

plt.show()
```
